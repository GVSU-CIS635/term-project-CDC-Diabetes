```python
# Importing Libraries
import pandas as pd

# Read the raw data file
raw_data = pd.read_csv("/content/drive/MyDrive/Project Data/2015.csv")
display(raw_data.head(10))
```

Selecting and renaming the features based on the BRFSS 2015 dataset

```python
# Sectioning the dataset to focus on the 23 variables of our selected refined dataset
cdc_indicators = raw_data[['DIABETE3',
                                         '_RFHYPE5',
                                         'TOLDHI2', '_CHOLCHK',
                                         '_BMI5',
                                         'SMOKE100',
                                         'CVDSTRK3', '_MICHD',
                                         '_TOTINDA',
                                         '_FRTLT1', '_VEGLT1',
                                         '_RFDRHV5',
                                         'HLTHPLN1', 'MEDCOST',
                                         'GENHLTH', 'MENTHLTH', 'PHYSHLTH', 'DIFFWALK',
                                         'SEX', '_AGEG5YR', 'EDUCA', 'INCOME2' ]]

# Renaming the 23 variables
cdc_indicators = cdc_indicators.rename(columns = {'DIABETE3':'Diabetes_012',
                                         '_RFHYPE5':'HighBP',
                                         'TOLDHI2':'HighChol', '_CHOLCHK':'CholCheck',
                                         '_BMI5':'BMI',
                                         'SMOKE100':'Smoker',
                                         'CVDSTRK3':'Stroke', '_MICHD':'HeartDiseaseorAttack',
                                         '_TOTINDA':'PhysActivity',
                                         '_FRTLT1':'Fruits', '_VEGLT1':"Veggies",
                                         '_RFDRHV5':'HvyAlcoholConsump',
                                         'HLTHPLN1':'AnyHealthcare', 'MEDCOST':'NoDocbcCost',
                                         'GENHLTH':'GenHlth', 'MENTHLTH':'MentHlth', 'PHYSHLTH':'PhysHlth', 'DIFFWALK':'DiffWalk',
                                         'SEX':'Sex', '_AGEG5YR':'Age', 'EDUCA':'Education', 'INCOME2':'Income' })

display(cdc_indicators)
```

Data Preprocessing

```python
# Create the dataframe 441456 rows
cdc_frame = pd.DataFrame(cdc_indicators)

# DROP ROWS WITH MISSING VALUES
cdc_frame= cdc_frame.dropna()
# Look up results: ~100K rows eliminated - new total 343606 rows
display(cdc_frame)


# REMOVING DUPLICATED ROWS
cdc_frame = cdc_frame.drop_duplicates()
# Look up results: ~4000 rows eliminated - new total 339107 rows
display(cdc_frame)
```

```python
# SELECTING 5 FEATURES FOR OUR STUDY
cdc_five = cdc_frame[['Diabetes_012','HighChol','Smoker','MentHlth','Education','Income']]

display(cdc_five)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# LOOKING FOR OUTLIERS for DIABETES_012

# Box Plot for the target variable
sns.boxplot(data=cdc_five['Diabetes_012']).set(title='BoxPlots for cdc_five')
```

**RESULTS:**

These are the different values for Diabetes_012. We  are more interested into values that provide clear answers for the sake of the study.
* 1 Yes
* 2 Yes, but female told only during pregnancy
* 3 No
* 4 No, pre-diabetes or borderline diabetes
* 7 Don’t know/Not Sure
* 9 Refused

The values 7 and 9 are considered not relevant for the dataset as we can't assume their true value.



```python
# Removing rows that contain 7 and 9 values for Diabetes_012
outliers=[7,9]
cdc_five = cdc_five[cdc_five.Diabetes_012.isin(outliers) == False]
display(cdc_five)
```

**RESULTS AFTER DROPPING IRRELEVANT VALUES:**

We now have 4 values left for Diabetes_012. Based on the BRFSS 2015 dataset, we can bin the values into 3 categories

Category 0: No Diabetes (values 2 and 3)

Category 1: Prediabetes (value 4)

Category 2: Diabetes (value 1)

Reference
* 1 Yes
* 2 Yes, but female told only during pregnancy
* 3 No
* 4 No, pre-diabetes or borderline diabetes

The value 2 is not considered as a chronic diabetes as it happened only during pregnancy.

```python
# Replacing the values with the 3 categories 0, 1, and 2

# Category 0
cdc_five['Diabetes_012']= cdc_five['Diabetes_012'].replace(2,0)
cdc_five['Diabetes_012']= cdc_five['Diabetes_012'].replace(3,0)

# Category 2
cdc_five['Diabetes_012']= cdc_five['Diabetes_012'].replace(1,2)

# Category 1
cdc_five['Diabetes_012']= cdc_five['Diabetes_012'].replace(4,1)

display(cdc_five)

# Bar Chart to see representation by Diabetes results

cdc_five.groupby(['Diabetes_012']).size().plot(kind = "bar")
```

```python
# OUTLIERS FOR VARIABLE MENTHLTH

# Box Plot
sns.boxplot(data=cdc_five['MentHlth']).set(title='BoxPlots for cdc_five')

```
